# Haaracascade Classifier

만약 Haaracascade Classifier에 대한 추가적인 정보가 필요하시다면 [Cascade 분류기](../0_thoery/2_Cascade_%EB%B6%84%EB%A5%98%EA%B8%B0.md)를 참고해주세요.

> cv2.CascadeClassifier.detectMultiScale의 추가인자 4종을 순서에 따라서 튜닝하세요
> 
> 1. scaleFactor
> 2. minNeighbors
> 3. minSize
> 4. maxSize

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
org_car = cv2.imread('../../assets/Images/clock.jpg')
cvt_car = cv2.cvtColor(org_car, cv2.COLOR_BGR2RGB)
gray_car = cv2.cvtColor(org_car, cv2.COLOR_BGR2GRAY)

clock_detector = cv2.CascadeClassifier('../../assets/Cascades/clocks.xml')
clock_detections = clock_detector.detectMultiScale(gray_car, scaleFactor=1.0019, minNeighbors=10, maxSize=(200, 200))

for (x, y, w, h) in clock_detections:
    print(w, h)
    cv2.rectangle(cvt_car, (x, y), (x+w, y+h), (255, 0, 0), 5)

plt.imshow(cvt_car)
```
