# Histograms of Oriented Gradients

만약 HOG Classifier에 대한 추가적인 정보가 필요하시다면 [HOG 분류기](../0_thoery/3_HOG_%EB%B6%84%EB%A5%98%EA%B8%B0.md)를 참고해주세요.

```python
import cv2
import dlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
org_img = cv2.imread('../../assets/Images/people2.jpg')
cvt_img = cv2.cvtColor(org_img, cv2.COLOR_BGR2RGB)

face_detector = dlib.get_frontal_face_detector()
face_detections = face_detector(cvt_img, 1)

for face in face_detections:
    pt1, pt2 = (face.left(), face.top()), (face.right(), face.bottom())    
    cv2.rectangle(cvt_img, pt1, pt2, (255, 0, 0), 3)

plt.imshow(cvt_img)
```
